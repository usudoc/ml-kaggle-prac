# Part2
## データの整形 1

```
game1_df = pd.read_csv("game1.csv")

game1_df.head()
```

column名の変更

```
game1_df = game1_df.rename(columns={"cellphone": "device_type"})
```

欠損地のある行の削除

```
game1_df = game1_df.dropna()
game1_df.to_csv("game1_formatted.csv", index=False)
```

## データ整形 2

```
game2_df = pd.read_csv("game2.csv")

game2_df.head()
game2_df.columns
```

ranking columnの削除

```
game2_df = game2_df.drop("ranking", axis=1)

game2_df.head()
```

欠損値を0で埋める

```
game2_df = game2_df.fillna(0)

game2_df.head()
```

columnで降順に並び替え

```
game2_df = game2_df.sort_values("billing_amount(¥)", ascending=False).reset_index(drop=True)
game2_df.to_csv("game2_formatted.csv", index=False)
```

# Part3

データの連結

```
kanto_df = pd.read_csv("kanto.csv")

kanto_df.head()
```

```
kansai_df = pd.read_csv("kansai.csv")

kansai_df.head()
```

```
kanto_kansai_df = pd.concat([kanto_df, kansai_df], sort=False)
kanto_kansai_df = kanto_kansai_df.reset_index(drop=True)

kanto_kansai_df.to_csv("kanto_kansai.csv", index=False)
kanto_kansai_df
```

データの結合

```
class_df = pd.read_csv("class.csv")

class_df
```

```
grade_df = pd.read_csv("grade.csv")

grade_df
```

```
class_grade_df = pd.merge(class_df, grade_df, on=["student_number", "name"], how="inner")

class_grade_df.to_csv("class_grade.csv", index=False)
class_grade_df
```
